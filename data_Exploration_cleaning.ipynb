```python
# Step 1 — First identify missing values BEFORE filling
print("Missing values per column:")
print(csv_1.isnull().sum())

# Step 2 — Then fill
numeric_cols = csv_1.select_dtypes(include="number").columns
for col in numeric_cols:
    csv_1[col] = csv_1[col].fillna(csv_1[col].median())

text_cols = csv_1.select_dtypes(include=["object", "string"]).columns
for col in text_cols:
    csv_1[col] = csv_1[col].fillna("Not Available")

# Step 3 — Then print AFTER filling 
print(f"Total missing values after filling: {csv_1.isnull().sum().sum()}")
```

```python
filtered_csv_1 = csv_1[csv_1["rating"] > 4]
print(f"\nHigh-rated products (rating > 4): {len(filtered_csv_1)} records")
print(filtered_csv_1.head())
```

```python
csv_1[["title", "rating"]].head()
```

```python
print(f"\nDuplicate rows found: {csv_1.duplicated().sum()}")
csv_1 = csv_1.drop_duplicates()
csv_1 = csv_1.reset_index(drop=True)
print(f"Duplicates after removal: {csv_1.duplicated().sum()}")
```

```python
csv_1["final_price"] = (
    csv_1["final_price"]
    .astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.replace('"', "", regex=False)
    .astype(float)
)
```

```python
csv_1["quantity"] = 1
```

```python
csv_1["total_amount"] = csv_1["final_price"] * csv_1["quantity"]
```

```python
csv_1[["final_price", "quantity", "total_amount"]].head()
```

```python
csv_1.to_csv("cleaned_dataset.csv", index=False)
print("\nCleaned dataset saved as 'cleaned_dataset.csv'")
print("Final Shape: {csv_1.shape[0]} rows x {csv_1.shape[1]} columns")
```
