```python
import pandas as pd 
import sqlite3
```

## Step 1: Load Dataset

```python
df = pd.read_csv(
    "../data/Sample_Superstore.csv",
    encoding="latin1"
)

df.head()
```

## Step 2: Load Dataset into SQL Database

```python
conn = sqlite3.connect("superstore.db")

df.to_sql(
    "sample_superstore",
    conn,
    if_exists="replace",
    index=False
)

print("Dataset loaded into SQL database successfully.")
```

## Step 3: Explore Table Structure and Sample Data

```python
query = """
SELECT * 
FROM sample_superstore
LIMIT 10;
"""

pd.read_sql(query,conn)
```

## Step 4: Apply WHERE Filters

```python
query = """
SELECT *
FROM sample_superstore
WHERE Region = 'West';
"""

pd.read_sql(query,conn)
```

### Filter Records by Category

```python
query = """
SELECT *
FROM sample_superstore
WHERE Category = 'Technology';
"""

pd.read_sql(query, conn)
```

### Filter Records by Sales Value

```python
query = """ 
SELECT *
FROM sample_superstore
WHERE Sales > 500;
"""

pd.read_sql(query,conn)
```

```python
df['Order Date'].head()
```

### Filter Records by Year (2017)

```python
query = """
SELECT *
FROM sample_superstore
WHERE [Order Date] LIKE '%2017';
"""

pd.read_sql(query, conn)
```

## Step 5: Aggregation using GROUP BY

```python
query = """
SELECT Region,
SUM(Sales) AS TotalSales
FROM sample_superstore
GROUP BY Region;
"""

pd.read_sql(query, conn)
```

```python
query = """ 
SELECT Category, 
SUM(Quantity) AS TotalQuantity
FROM sample_superstore
GROUP BY Category;
"""

pd.read_sql(query,conn)
```

```python
query = """ 
SELECT Category,
AVG(Sales) AS AverageSales
FROM sample_superstore
GROUP BY Category;
"""

pd.read_sql(query,conn)
```

```python
df.columns.tolist()
```

## Step 6: Sort and Limit Results

```python
query =""" 
SELECT [Product Name],
SUM(Sales) AS TotalSales
FROM sample_superstore
GROUP BY [Product Name]
ORDER BY TotalSales DESC
LIMIT 10;
"""

pd.read_sql(query,conn)
```

```python
query = """
SELECT Category,
SUM(Sales) AS TotalSales
FROM sample_superstore
GROUP BY Category
ORDER BY TotalSales DESC;
"""

pd.read_sql(query, conn)
```

## Step 7: Business Use Cases

```python
df['Order Date'] = pd.to_datetime(
    df['Order Date'],
    format='mixed'
)
```

```python
df['Order Date'].head()
```

```python
monthly_sales = (
    df.groupby(df['Order Date'].dt.month)['Sales']
    .sum()
    .reset_index()
)

monthly_sales
```

### Monthly Sales Trend Analysis

Monthly sales trend was analyzed by extracting month information from the Order Date column and aggregating total sales month-wise.

### Top Customers by Sales

```python
query = """ 
SELECT [Customer Name],
SUM(Sales) AS TotalSales
FROM sample_superstore
GROUP BY [Customer Name]
ORDER BY TotalSales DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

### Duplicate Orders

```python
query = """
SELECT [Order ID],
COUNT(*) AS DuplicateCount
FROM sample_superstore
GROUP BY [Order ID]
HAVING COUNT(*) > 1;
"""

pd.read_sql(query, conn)
```
