## Introduction to CTE

A Common Table Expression (CTE) is a temporary result set created using the WITH keyword.

It exists only during the execution of a single SQL statement.

CTEs make long SQL queries easier to understand and improve query readability.

## Real-World Example

Suppose a business analyst needs to calculate the total sales of every customer and use the same result multiple times in a query.

Instead of writing the same calculation repeatedly, a Common Table Expression can create a temporary result set that makes the query simpler and easier to understand.

## Why do we use CTEs?

CTEs help simplify complex queries.

They improve readability, reduce repetition and make SQL code easier to maintain.

Instead of writing nested queries multiple times, we can create a temporary table using a CTE and reuse it.

## Query 3

### Problem Statement

Calculate the total sales for each customer.

### Understanding the Problem

A customer may place multiple orders.

To know the total contribution of each customer, we need to add the sales of all orders placed by that customer.

A Common Table Expression is used to calculate the total sales.

```python
# Query 3
# Calculate total sales for each customer using CTE. 

query3 = """

WITH customer_sales AS
(
SELECT 
customer_id,
SUM(sales) AS total_sales
FROM orders
GROUP BY customer_id
)
SELECT * 
FROM customer_sales
ORDER BY total_sales DESC;
"""

pd.read_sql(query3, conn)
```

### Observation

The query calculates the cumulative sales generated by every customer.

The result provides a summarized view of customer contributions and helps identify customers with higher overall sales.

## Query 4
### Problem Statement

Find customer whose total sales are above average.

### Understanding the Problem 

First, calculate the total sales of each customer.

Then calculate the average of those total sales values.

Display only those customers whose total sales are greater than the average.

```python
# Query 4
# Find customers whose total sales are above average 

query4 = """

WITH customer_sales AS

(
SELECT 
customer_id,
SUM(sales) AS total_sales
FROM orders
GROUP BY customer_id
)

SELECT *
FROM customer_sales
WHERE total_sales >
(
SELECT AVG(total_sales)
FROM customer_sales
);
"""
pd.read_sql(query4, conn)
```

### Observation

The query identifies customers whose cumulative sales exceed the average total sales across all customers.

These customers represent a high-performing customer segment within the dataset.
