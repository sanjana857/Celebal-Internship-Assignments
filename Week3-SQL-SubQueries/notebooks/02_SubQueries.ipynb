## Introduction to Subqueries 

A subquery is a query written inside another SQL query.

The inner query executes first and returns a result.

The outer query then uses that result to produce the final output.

Subqueries are commonly used with WHERE, FROM and SELECT clauses.

## Real-World Example

Suppose a company wants to identify products priced above the average price.

Instead of calculating the average manually, SQL can calculate the average automatically and compare each product with that value using a subquery.

This makes data analysis faster and more efficient.

## Why do we use Subqueries?

Subqueries make SQL queries more flexible and dynamic.

Instead of manually calculating a value and using it in another query, SQL performs the calculation automatically.

This makes the query easier to maintain and reduces manual work.

## Types of Subqueries

Subqueries can be classified into different types based on their usage.

- Single-row Subquery: Returns only one value.
- Multiple-row Subquery: Returns multiple values.
- Correlated Subquery: Depends on the outer query for execution.

In this notebook, Query 1 uses a single-row subquery and Query 2 uses a correlated subquery.

## QUERY 1

### Problem Statement

Find all orders where sales are greater than the average sales.

### Understanding the Problem

First, we need to calculate the average sales value from the orders table.

After that, we compare each order with the average sales.

Only those orders whose sales are greater than the average value will be displayed.

```python
# Query 1
# Display all orders where sales are greater than the average sales

query1 = """

SELECT
    order_id,
    customer_id,
    sales,
    quantity,
    profit

FROM orders
WHERE sales >
(
SELECT AVG(sales)
FROM orders
);
"""
pd.read_sql(query1, conn)
```

### Observation

The query displays only those orders whose sales value is greater than the average sales across all orders.

These orders represent higher-value transactions and can help identify significant sales activities.

## Query 2

### Problem Statement

Find the highest sales order for each customer.

### Understanding the Problem

A customer may place multiple orders.

The objective is to identify the order with the highest sales value for every customer.

A correlated subquery is used to compare each customer's sales with their maximum sales amount.

```python
# Query 2
# Find the highest sales order for each customer.

# Display order details

query2 = """

SELECT 
    order_id,
    customer_id,
    sales,
    quantity,
    profit

FROM orders o

WHERE sales = 

(

SELECT MAX(sales)

FROM orders

WHERE customer_id = o.customer_id
);
"""

pd.read_sql(query2, conn)
```

### Observation

The query identifies the order with the highest sales value for every customer.

If multiple orders have the same maximum sales value, all matching records are included in the output.
