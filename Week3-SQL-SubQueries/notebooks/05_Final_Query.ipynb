## Problem Statement

Display the following information:

- Customer Name
- Total Sales
- Customer Rank

The solution should use JOIN, CTE and Window Function together.

## Approach

1. Calculate total sales for every customer using a CTE.

2. Join the calculated sales with the customers table.

3. Rank customers according to their total sales using the RANK() window function.

4. Display the final report.

```python
# Final QUery
# Display customer name, total sales and rank

query = """
WITH customer_sales AS
(
SELECT
customer_id,
SUM(sales) AS total_sales
FROM orders
GROUP BY customer_id
)

SELECT 
c.customer_name,
cs.total_sales,
RANK()
OVER(
ORDER BY cs.total_sales DESC
)

AS customer_rank
FROM customer_sales cs
JOIN customers c
ON cs.customer_id = c.customer_id;
"""

pd.read_sql(query, conn)
```

### Observation

The query successfully combines customer information with sales data.

Each customer is assigned a rank based on total sales, making it easy to compare customer performance and identify the highest revenue-generating customers.

Customers with higher sales receive better ranks.

The report combines information from multiple tables in a single result.
